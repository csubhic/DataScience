```
# On the same boxplot, highlight the difference between having central air conditioning or not.

import pandas as pd
import seaborn as sns
sns.set()

houseprices_df = pd.read_csv('https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/houseprices.csv')
sns.set_style('white')
box_overallcond_x_y = sns.boxplot(x="overallcond",y="saleprice",hue="centralair",data=houseprices_df);
```
