```
# reverse a string
def reverse(inp):
  revstr = ""
  if len(inp) == 0:
    print('Enter a string to reverse')
    return inp
  else:
   for i in inp:
    revstr = i + revstr
  return revstr

print('Reversed String :'+reverse('Thinkful'))
```

```
#Baye's theorem
# P(A/B) = P(B/A)(PA)
#          ---------
#            P(B)
#P(B)=P(B/A)P(A)+P(B/~A)P(~A)
def bayes(a,b_given_a,b_given_not_a):
    return((b_given_a * a) / (b_given_a* a )+ (b_given_not_a)*(1-a))

bayes(-2,-5,-3)
```
