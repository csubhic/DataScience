```
#Law of Large numbers 
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from numpy import mean
from scipy import stats
from numpy.random import randint

results = []

for num_trials in range(1,1000):
    trials = np.random.normal(1,7,num_trials)
    mean_of_trials = trials.mean()
    results.append(mean_of_trials)

df=pd.DataFrame({'trials':results})
df.plot()
```

```
#central limit theorem - Normal Distribution
means = [mean(np.random.normal(1,7,1000)) for _ in range(1000)]

plt.hist(means)
plt.show()
```

```
#Law of large numbers = Gamma

results = []

for num_trials in range(1,1000):
    trials = np.random.gamma(4,1,num_trials)
    mean_trials = trials.mean()
    results.append(mean_trials)

df=pd.DataFrame({'trials':results})
df.plot()
```

```
#central limit theorem - gamma

means = [mean(np.random.gamma(4,1,1000)) for _ in range(1,5000)]
plt.hist(means)
plt.show()
```
