```
#normal
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

rand1= np.random.normal(5,.5,100)
rand1.sort()
plt.hist(rand1,bins=20,color="c")

plt.axvline(rand1.mean(), color="b",linestyle="solid",linewidth=2)
plt.axvline(rand1.mean()+rand1.std(),color="b",linestyle="dashed",linewidth=2)
plt.axvline(rand1.mean()-rand1.std(),color="b",linestyle="dashed",linewidth=2)
plt.show()
```

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

rand1=np.random.normal(5,.5,100)
rand2=np.random.normal(10,1,100)

rand3=rand1 + rand2
rand3_mean = np.mean(rand3)
rand3_std = np.std(rand3)
plt.hist(rand3,bins=20,color="c")
plt.axvline(rand3_mean,color="b",linestyle="solid",linewidth=2)
plt.axvline(rand3_mean + rand3_std, color="b",linestyle="dashed",linewidth=2)
plt.axvline(rand3_mean - rand3_std, color="b",linestyle="dashed",linewidth=2)
plt.show()
```
