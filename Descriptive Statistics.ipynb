```
import pandas as pd
from scipy import stats
players_df = pd.read_csv("https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/players.csv")
players_df.info()

players_1990 = players_df.iloc[(players_df['yearid'] == 1990).values]
players_1990.info()

players_2000 = players_df.iloc[(players_df['yearid'] == 2000).values ]
players_2000.info()

stats.ttest_ind(players_2000['salary'],players_1990['salary'])
 
```

```
#Is there a significant correlation between the lotarea and saleprice variables? Are they positively or negatively correlated?
import pandas as pd
from scipy import stats
houseprices_df=pd.read_csv('https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/houseprices.csv')
salepricerange = houseprices_df['saleprice'].describe()
lot_area_range = houseprices_df['lotarea'].describe()
corr_sale_lot = salepricerange.corr(lot_area_range)
print("Correlation between Sale Price and Lot Area range is:\n {}".format(corr_sale_lot) +' ' + "and it is positively correlated")
```
